```python
from keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = load_model('food_prediction.model')

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
img = cv2.imread('halwa.jpg')
img = cv2.resize(img,(120,120))
image=img
plt.imshow(img)
plt.show()
img = np.reshape(img,[1,120,120,3])
classes = model.predict_classes(img)
a=classes[0]
catagories=["biriyani","bisibelebath","butternaan","chaat","chappati","dhokla","dosa","gulab jamun","halwa","idly","kathi roll","meduvadai","noodles","paniyaram","poori","samosa","tandoori chicken","upma","vada pav","ven pongal"]
print(catagories[a])
```
